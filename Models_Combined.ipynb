```python
import pandas as pd
import numpy as np
import joblib
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import HistGradientBoostingRegressor
import matplotlib.pyplot as plt
from sklearn.inspection import PartialDependenceDisplay, permutation_importance
import seaborn as sns
from sklearn.compose import TransformedTargetRegressor
from sklearn.preprocessing import QuantileTransformer
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV


#Load cleaned data
df = pd.read_csv('steam_ml_cleaned_data.csv').dropna()

#X: features (skipping first 3), y: target
X = df.iloc[:, 3:-1].copy()
y = df.iloc[:, -1]

#Split into training and testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1
)

#Define scoring metrics 
scoring = {
    'R2': 'r2',
    'MAE': 'neg_mean_absolute_error',
    'MSE': 'neg_mean_squared_error'
}


"""
Histboost Regressor Training
"""
#Create histboost regressor with early stopping
hgb = HistGradientBoostingRegressor(
    random_state=1,
    early_stopping=True,
    n_iter_no_change=15,
    validation_fraction=0.1
)

#Create parameter distribution to use with search for HGBR
param_distributions = {
    'learning_rate': [0.01, 0.05, 0.1],
    'max_iter': [100, 300, 500],
    'max_depth': [3, 5, 10, None],
    'l2_regularization': [0, 0.1, 1.0],
    'min_samples_leaf': [10, 25, 50]
}

#Implement random search for histboost regressor with parameter distribution
# and five fold cross validation using scoring metrics with refit metric r2
random_search = RandomizedSearchCV(
    estimator=hgb,
    param_distributions=param_distributions,
    n_iter=40,                       
    scoring=scoring,
    cv=5,
    refit='R2',
    n_jobs=-1,
    verbose=1,
    random_state=21
)

#Train the model
random_search.fit(X_train, y_train)

best_model = random_search.best_estimator_
y_pred = best_model.predict(X_test)


"""
Random Forest Training
"""
#Random Forest grid for training
rf_param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'max_features': ['sqrt', 'log2']
}

#Grid search for Random Forest
grid_search_rf = GridSearchCV(
    estimator=RandomForestRegressor(random_state=1, n_jobs=-1),
    param_grid=rf_param_grid,
    cv=3, 
    scoring=scoring,
    refit='R2',
    verbose=1
)

#Do the grid search
grid_search_rf.fit(X_train, y_train)

#Find best model
best_rf_model = grid_search_rf.best_estimator_
y_pred_rf = best_rf_model.predict(X_test)


"""
Linear Regression Training
"""
#Linear Regression grid for training
lr_param_grid = {
    'fit_intercept': [True, False],
    'positive': [True, False], 
    'copy_X': [True, False],    
}

lr_grid = GridSearchCV(LinearRegression(), lr_param_grid, cv=5, scoring=scoring, refit='R2')

#Train models
lr_grid.fit(X_train, y_train)

#Find best model
best_lr = lr_grid.best_estimator_


"""
Plotting Methods
"""
#Display top trained models and data associated with them
def display_search_results(search_object, top_n=5):
    results_df = pd.DataFrame(search_object.cv_results_)
    
    relevant_columns = ['params', 'rank_test_R2']
    score_cols = [
        'mean_test_R2', 'std_test_R2', 
        'mean_test_MAE', 'std_test_MAE', 
        'mean_test_MSE', 'std_test_MSE'
    ]
    
    final_columns = relevant_columns + score_cols
    
    existing_cols = [col for col in final_columns if col in results_df.columns]
    
    #Sort by R2 rank and display
    display_table = results_df[existing_cols].sort_values('rank_test_R2')
    
    print(f"Top {top_n} Parameter Combinations:")
    pd.set_option('display.max_colwidth', None)
    display(display_table.head(top_n))

#Residuals Plot
def plot_residuals(model, X_test, y_test, model_name="Model"):
    y_pred = model.predict(X_test)
    residuals = y_test - y_pred
    
    plt.figure(figsize=(8, 6))
    plt.scatter(y_pred, residuals, alpha=0.5, edgecolors='k')
    plt.axhline(0, color='red', linestyle='--')
    plt.xlabel('Predicted Values')
    plt.ylabel('Residuals/Errors (Actual - Predicted)')
    plt.title(f'{model_name} Predicted vs. Residuals Plot')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

#Plot both error distribution and hexbin plots
def plot_prediction_analysis(y_true, y_pred, model_name="Model"):
    #error distribution plot
    residuals = y_true - y_pred
    plt.figure(figsize=(8, 4))
    sns.histplot(residuals, kde=True, color='purple')
    plt.axvline(0, color='black', linestyle='--')
    plt.title(f'{model_name}: Error Distribution')
    plt.xlabel('Residual/Error (Actual - Predicted)')
    plt.show()

    #hexbin plot
    plt.figure(figsize=(10, 7))
    hb = plt.hexbin(y_true, y_pred, gridsize=30, cmap='YlGnBu', mincnt=1)
    plt.colorbar(hb, label='Number of Games')
    plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--', lw=2)
    plt.title(f'{model_name}: Prediction Density')
    plt.xlabel('Actual Positive Review Ratio')
    plt.ylabel('Predicted Positive Review Ratio')
    plt.show()

#Plot top 4 most important features and their impact on the model
def plot_importance(model, X_test, y_test, feature_names, top_n=4, title="Top Positive Feature Impact"):
    perm_importance = permutation_importance(model, X_test, y_test, n_repeats=10, random_state=7)
    importances = pd.Series(perm_importance.importances_mean, index=feature_names)
    sorted_importances = importances.sort_values()
    df_top = sorted_importances.tail(top_n)
    plt.figure(figsize=(10, 5))
    df_top.plot(
        kind='barh', 
        color='skyblue',
        edgecolor='black'
    )
    plt.title(f'Top {top_n} Permutation Importances (Test Set)')
    plt.xlabel("Decrease in R2 Score when Permuted")
    plt.tight_layout()
    plt.show()
    return df_top.sort_values(ascending=False).index.tolist()
    
#Create partial dependence plots for top features
def create_partial_dependence_plot_top_features(top_features, model, X_train, title):
    fig, axes = plt.subplots(1, 4, figsize=(20, 5), sharey=True)
    
    display1d = PartialDependenceDisplay.from_estimator(
        model, 
        X_train, 
        features=top_features,
        feature_names=X.columns.tolist(), 
        ax=axes,
        grid_resolution=50,
        line_kw={"color": "tab:blue", "linewidth": 2}
    )
    # plot partial dependence plots with labels
    for i, ax in enumerate(axes):
        ax.set_title(f'Effect of {top_features[i]}', fontsize=12, fontweight='bold')
        ax.set_ylabel('Change in Predicted Review Ratio' if i==0 else "")
        ax.grid(True, linestyle='--', alpha=0.6)
        ax.axhline(0, color='black', linewidth=0.8, alpha=0.5)
    plt.suptitle(title, fontsize=16, y=1.05)
    plt.tight_layout()
    plt.show()

# Plotting the distribution of the target variable (y_train)
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(y, kde=True, color='purple', bins=30)
plt.title('Distribution of Positive Review Ratio', fontsize=15)
plt.xlabel('Positive Review Ratio', fontsize=12)
plt.ylabel('Frequency (Number of Games)', fontsize=12)
plt.axvline(y.mean(), color='red', linestyle='--', label=f'Mean: {y.mean():.2f}')
plt.legend()
plt.show()


"""
Histboost Regressor Graphs
"""
print("\n\n----- Histboost Regressor -----\n")

#Best Perams
display_search_results(random_search)

#Residuals plot
plot_residuals(best_model, X_test, y_test, "HistBoost Regressor")

#Error Distribution and Hexbin plot
plot_prediction_analysis(y_test, y_pred, "Histogram Boosting Regression")

perm_importance.feature_importances_ = perm_importance.importances_mean

#Top four features plot
top_4_features = plot_importance(best_model, X_test, y_test, X_train.columns, top_n=4, title="Optimized HGBR: Permutation Importance")

#Four graphs to show how top 4 features individually influence 
create_partial_dependence_plot_top_features(top_4_features, best_model, X_train,
                            'HGBR: How the Top 4 Features Individually Influence Review Ratio')


"""
Random Forest Plots 
"""
print("\n\n----- Random Forest -----\n")

#Best Perams
display_search_results(grid_search_rf)

#Residuals plot
plot_residuals(grid_search_rf.best_estimator_, X_test, y_test, "Random Forest")

#Error Distribution and Hexbin plot
plot_prediction_analysis(y_test, y_pred_rf, "Optimized Random Forest")

#Top four features plot
rf_top_4 = plot_importance(best_rf_model, X_test, y_test, X_train.columns, top_n=4, title="Optimized RF: Top Features")

#Four graphs to show how top 4 features individually influence 
create_partial_dependence_plot_top_features(rf_top_4, best_rf_model, X_train,
                            'HGBR: How the Top 4 Features Individually Influence Review Ratio')


"""
Linear Regression Plots 
"""
print("\n\n----- Linear Regression -----\n")

#Best Perams
display_search_results(lr_grid)

#Residuals plot
plot_residuals(lr_grid.best_estimator_, X_test, y_test, "Linear Regression")

#Error Distribution and Hexbin plot
plot_prediction_analysis(y_test, y_pred_lr, "Optimized Linear Regression")

#Top four features plot
lr_top_4 = plot_importance(best_lr, X_test, y_test, X_train.columns, title="Best Linear: Feature Impact")

#Four graphs to show how top 4 features individually influence 
create_partial_dependence_plot_top_features(lr_top_4, best_lr, X_train,
                            'HGBR: How the Top 4 Features Individually Influence Review Ratio')
```
