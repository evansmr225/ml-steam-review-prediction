```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
import seaborn as sns
from sklearn.metrics import r2_score

# Load cleaned data
df = pd.read_csv('steam_ml_cleaned_data.csv').dropna()

# X: features (skipping first 3), y: target
X = df.iloc[:, 3:-1].copy()
y = df.iloc[:, -1]

# split into training and testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1
)


#helper functions for graphs
def plot_importance(model, feature_names, top_n=4, title="Top Positive Feature Impact"):
    """
    Displays top features in a bar graph
    """
    plt.figure(figsize=(10, 4))
    
    #get values based on model type
    if hasattr(model, 'feature_importances_'):
        vals = model.feature_importances_
    else:
        vals = model.coef_

    #create DataFrame
    df = pd.DataFrame({'Feature': feature_names, 'Value': vals})

    #get the largest values
    df_top = df.sort_values(by='Value', ascending=False).head(top_n)

    #plot
    plt.barh(df_top['Feature'], df_top['Value'], color='skyblue')
    plt.gca().invert_yaxis()
    plt.title(title)
    plt.xlabel('Impact Score (Positive Only)')
    plt.show()
    
def plot_prediction_analysis(y_true, y_pred, model_name="Model"):
    """
    Hexbin density plot and the Residual distribution.
    """
    #hexbin plot
    plt.figure(figsize=(10, 7))
    hb = plt.hexbin(y_true, y_pred, gridsize=30, cmap='YlGnBu', mincnt=1)
    plt.colorbar(hb, label='Number of Games')
    plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--', lw=2)
    plt.title(f'{model_name}: Prediction Density')
    plt.xlabel('Actual Positive Review %')
    plt.ylabel('Predicted Positive Review %')
    plt.show()

    #residual plot
    residuals = y_true - y_pred
    plt.figure(figsize=(8, 4))
    sns.histplot(residuals, kde=True, color='purple')
    plt.axvline(0, color='black', linestyle='--')
    plt.title(f'{model_name}: Error Distribution')
    plt.xlabel('Error (Actual - Predicted)')
    plt.show()

#Do random forest Regression
rf_model = RandomForestRegressor(n_estimators=50, random_state=1, n_jobs=-1)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)

#R2 value for random forest - 0.2769%
print(f"Random Forest Test R2: {rf_model.score(X_test, y_test):.4f}")

#plot random forest graphs
plot_importance(rf_model, X_train.columns, top_n=4, title="RF: Top 4 Features")
plot_prediction_analysis(y_test, y_pred_rf, "Random Forest")


#linear regression
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
y_pred_lr = lr_model.predict(X_test)

#R2 test for linear regression - 0.1396%
print(f"Linear Regression Test R2: {lr_model.score(X_test, y_test):.4f}")

#plot linear regression graphs
plot_importance(lr_model, X_train.columns, top_n=4, title="Linear Reg: Impact Features")
plot_prediction_analysis(y_test, y_pred_lr, "Linear Regression")
```
